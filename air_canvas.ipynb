```python
# part 2 capture window and track landmarks and draw connectivites btwn landmarks
import cv2
import mediapipe as mp
import numpy as np 
mphand=mp.solutions.hands
hands=mphand.Hands()

mpDraw=mp.solutions.drawing_utils

cap=cv2.VideoCapture(0)

while True:
        ret, frame = cap.read()
        x, y, c = frame.shape

        # Image ko RGB format mein convert karna
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Hand landmarks ko detect karna
        results = hands.process(frame_rgb)

        if results.multi_hand_landmarks:
            for hand_landmarks in results.multi_hand_landmarks:
                # Haath ke landmarks ko draw karna
                for idx, landmark in enumerate(hand_landmarks.landmark):
                    cx, cy = int(landmark.x * frame.shape[1]), int(landmark.y * frame.shape[0])
                    cv2.circle(frame, (cx, cy), 5, (255, 0, 0), -1)

                # Landmarks ko connect karne ke liye
                mp_drawing = mp.solutions.drawing_utils
                mp_drawing.draw_landmarks(frame, hand_landmarks, mphand.HAND_CONNECTIONS)

        # Frame ko display karna
        cv2.imshow('Hand Landmarks', frame)

        # 'q' press karke loop se bahar nikalna
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Resources release karna
cap.release()
cv2.destroyAllWindows()
```

```python
# part 4  the logic inserting the tracked points in deques , and selection of color (in this logic we have 
# not drawn the points into window)
import cv2
import mediapipe as mp
import numpy as np
from collections import deque

# Initialize deques and indices for - blue,green,red,yellow colors
bpoints = [deque(maxlen=1024)]
gpoints = [deque(maxlen=1024)]
rpoints = [deque(maxlen=1024)]
ypoints = [deque(maxlen=1024)]

# These indices will be used to mark the points in particular deque of the specified color
blue_index = 0
green_index = 0
red_index = 0
yellow_index = 0

# to choose the colors we need colors list in RGB formate
colors = [(255, 0, 0), (0, 255, 0), (0, 0, 255), (0, 255, 255)]
colorIndex = 0



mphand=mp.solutions.hands
hands=mphand.Hands()

mpDraw=mp.solutions.drawing_utils

cap=cv2.VideoCapture(0)

while True:
        ret, frame = cap.read()
        x, y, c = frame.shape

        

        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        frame = cv2.rectangle(frame, (40, 1), (140, 65), (0, 0, 0), 2)
        frame = cv2.rectangle(frame, (160, 1), (255, 65), (255, 0, 0), 2)
        frame = cv2.rectangle(frame, (275, 1), (370, 65), (0, 255, 0), 2)
        frame = cv2.rectangle(frame, (390, 1), (485, 65), (0, 0, 255), 2) 
        frame = cv2.rectangle(frame, (505, 1), (600, 65), (0, 255, 255), 2)
        cv2.putText(frame, "CLEAR", (49, 33), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2, cv2.LINE_AA)
        cv2.putText(frame, "BLUE", (185, 33), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2, cv2.LINE_AA)
        cv2.putText(frame, "GREEN", (298, 33), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2, cv2.LINE_AA)
        cv2.putText(frame, "RED", (420, 33), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2, cv2.LINE_AA)
        cv2.putText(frame, "YELLOW", (520, 33), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2, cv2.LINE_AA)

        
        results = hands.process(frame_rgb)

        if results.multi_hand_landmarks:
            landmarks = []
            for hand_landmarks in results.multi_hand_landmarks:
               
                for idx, landmark in enumerate(hand_landmarks.landmark):
                    cx, cy = int(landmark.x * frame.shape[1]), int(landmark.y * frame.shape[0])
                    
                    landmarks.append([cx, cy])
                
                mp_drawing = mp.solutions.drawing_utils
                mp_drawing.draw_landmarks(frame, hand_landmarks, mphand.HAND_CONNECTIONS)
            fore_finger = (landmarks[8][0], landmarks[8][1])# Extracts the coordinates of the index finger landmark.
            center = fore_finger # Assigns the extracted index finger coordinates to the variable 'center'.
            thumb = (landmarks[4][0], landmarks[4][1]) # Extracts the coordinates of the thumb landmark.
            cv2.circle(frame, center, 3  , (0, 255, 0), -1)# Draws a filled circle representing the index finger on the frame.
            


            # if thumb and fore_finger is  joined
            if (thumb[1] - center[1] < 30):
                bpoints.append(deque(maxlen=512))
                blue_index += 1
                gpoints.append(deque(maxlen=512))
                green_index += 1
                rpoints.append(deque(maxlen=512))
                red_index += 1
                ypoints.append(deque(maxlen=512))
                yellow_index += 1
                
            #when center is in color panel 
            elif center[1] <= 65:
                if 40 <= center[0] <= 140:  # Clear Button
                    bpoints = [deque(maxlen=512)]
                    gpoints = [deque(maxlen=512)]
                    rpoints = [deque(maxlen=512)]
                    ypoints = [deque(maxlen=512)]

                    blue_index = 0
                    green_index = 0
                    red_index = 0
                    yellow_index = 0

                    paintWindow[ 67 :, :, : ] = 255
                elif 160 <= center[0] <= 255:
                    colorIndex = 0  # Blue
                elif 275 <= center[0] <= 370:
                    colorIndex = 1  # Green
                elif 390 <= center[0] <= 485:
                    colorIndex = 2  # Red
                elif 505 <= center[0] <= 600:
                    colorIndex = 3  # Yellow
            #when  we are drawing ,insert the tracked point into deque
            else:
                if colorIndex == 0:
                    bpoints[blue_index].appendleft(center)
                elif colorIndex == 1:
                    gpoints[green_index].appendleft(center)
                elif colorIndex == 2:
                    rpoints[red_index].appendleft(center)
                elif colorIndex == 3:
                    ypoints[yellow_index].appendleft(center)
    # Append the next deques when nothing is detected to avoid messing up
        else:
            bpoints.append(deque(maxlen=512))
            blue_index += 1
            gpoints.append(deque(maxlen=512))
            green_index += 1
            rpoints.append(deque(maxlen=512))
            red_index += 1
            ypoints.append(deque(maxlen=512))
            yellow_index += 1
        
        
        cv2.imshow('Hand Landmarks', frame)

        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    
cap.release()
cv2.destroyAllWindows()
```
